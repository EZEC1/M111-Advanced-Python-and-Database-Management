## Exercise 1
**Read Total profit of all months and show it using a line plot**

Total profit data provided for each month. Generated line plot must include the following properties:

- Y label name = Total profit
The line plot graph should look like this.

```python
# Write your code here
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv("../data/company_sales_data.csv")
f_data=np.array(data)
plt.plot(f_data[:,0], f_data[:,-1])
plt.xlabel("Month number")
plt.ylabel("Profil in dollar")
plt.title("Company profil per month")
plt.show()
```

## Exercise 3
**Read all product sales data and show it  using a multiline plot**
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).
**Expected output**

<img src="../data/sales_data.png" alt="sales_data" style="width: 500px;"/>


```python
# write your code here
data = pandas.read_csv(r"../data/company_sales_data.csv")
f_data=np.array(data)
plt.plot(f_data[:,0],f_data[:,1], color = 'blue',marker = '.',
         markerfacecolor = 'blue',label='facecream Sales Data')
plt.plot(f_data[:,0],f_data[:,2], color = 'orange',marker = '.',
         markerfacecolor = 'orange',label='facewash Sales Data ')
plt.plot(f_data[:,0],f_data[:,3], color = 'green',marker = '.',
         markerfacecolor = 'green',label='toothpaste Sales Data')
plt.plot(f_data[:,0],f_data[:,4], color = 'red',marker = '.',
         markerfacecolor = 'red',label='bathingsoap Sales Data')
plt.plot(f_data[:,0],f_data[:,5], color = 'purple',marker = '.',
         markerfacecolor = 'purple',label='shampoo Sales Data')
plt.plot(f_data[:,0],f_data[:,6], color = 'brown',marker = '.',
         markerfacecolor = 'brown',label='moisturizer Sales Data')
plt.xlabel("Month number")
plt.ylabel("Sales Units")
plt.title("Sales data")
plt.legend(loc="upper left")
plt.ylim(0,18000)
plt.show()
```

## Exercise 7
**Read the total profit of each month and show it using the histogram to see the most common profit ranges**

**Expected output**

<img src="../data/Profit_data.png" alt="Profit_data" style="width: 700px;"/>

## Exercise 8
**Calculate total sale data for last year for each product and show it using a Pie chart**

*Note: In Pie chart display Number of units sold per year for each product in percentage.*

**Expected output**

<img src="../data/sales_data_pi_chart.png" alt="sales_data_pi_chart" style="width: 700px;"/>


```python
# write your code here
month  = data ['month_number']
lab = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
salesData   = [data ['facecream'].sum(), data ['facewash'].sum(), data ['toothpaste'].sum(), 
         data ['bathingsoap'].sum(), data ['shampoo'].sum(), data ['moisturizer'].sum()]
plt.axis("equal")
plt.pie(salesData, labels=lab, autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.title('Sales data')
plt.show()
```

## Exercise 9
**Read Bathing soap facewash of all months and display it using the Subplot**

**Expected output**

<img src="../data/Sales_data_of_a_facewash.png" alt="Sales_data_of_a_facewash" style="width: 700px;"/>


```python
# write your code here
month  = data ['month_number']
bathingsoap   = data ['bathingsoap']
faceWashSalesData   = data ['facewash']

f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(month, bathingsoap, label = 'Bathingsoap Sales Data', color='k', marker='o', linewidth=3)
axarr[0].set_title('Sales data of  a Bathingsoap')
axarr[1].plot(month, faceWashSalesData, label = 'Face Wash Sales Data', color='r', marker='o', linewidth=3)
axarr[1].set_title('Sales data of  a facewash')

plt.xticks(month)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()
```

## Exercise 10
**Read all product sales data and show it using the stack plot**

**Expected output**

<img src="../data/product_sales_stack_plot.png" alt="product_sales_stack_plot" style="width: 700px;"/>


```python
# write your code here
faceCremSalesData   = data ['facecream']
faceWashSalesData   = data ['facewash']
toothPasteSalesData = data['toothpaste']
bathingsoapSalesData   = data ['bathingsoap']
shampooSalesData   = data ['shampoo']
moisturizerSalesData = data ['moisturizer']

plt.plot([],[],color='m', label='face Cream', linewidth=5)
plt.plot([],[],color='c', label='Face wash', linewidth=5)
plt.plot([],[],color='r', label='Tooth paste', linewidth=5)
plt.plot([],[],color='k', label='Bathing soap', linewidth=5)
plt.plot([],[],color='g', label='Shampoo', linewidth=5)
plt.plot([],[],color='y', label='Moisturizer', linewidth=5)

plt.stackplot(month, faceCremSalesData, faceWashSalesData, toothPasteSalesData, 
              bathingsoapSalesData, shampooSalesData, moisturizerSalesData, 
              colors=['m','c','r','k','g','y'])

plt.xlabel('Month Number')
plt.ylabel('Sales unints in Number')
plt.title('Alll product sales data using stack plot')
plt.legend(loc='upper left')
plt.ylim(0,40000)
plt.show()
```
